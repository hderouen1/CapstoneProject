```python
import pandas as pd
```

```python
#loading merged data
test = pd.read_csv('merged_test_dataset.csv')
train = pd.read_csv('merged_train_dataset.csv')
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder

# Step 1: Define features and target variable
X = train.drop(['Provider', 'PotentialFraud'], axis=1)  # Exclude Provider and PotentialFraud from features
y = train['PotentialFraud'].apply(lambda x: 1 if x == 'Yes' else 0)  # Encode 'Yes' as 1, 'No' as 0

# Step 2: Handle missing values
# Separate numeric and non-numeric columns
numeric_cols = X.select_dtypes(include=['number']).columns
non_numeric_cols = X.select_dtypes(exclude=['number']).columns

# Fill missing values for numeric columns with the mean
X[numeric_cols] = X[numeric_cols].fillna(X[numeric_cols].mean())

# Fill missing values for non-numeric columns with a placeholder
X[non_numeric_cols] = X[non_numeric_cols].fillna('missing')

# Step 3: Convert non-numeric columns to string to ensure uniformity
for col in non_numeric_cols:
    X[col] = X[col].astype(str)

# Step 4: Encode categorical features using LabelEncoder
label_encoders = {}
for col in non_numeric_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le

# Step 5: Split the dataset into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 6: Train a basic Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Step 7: Make predictions and evaluate the model
y_pred = rf_model.predict(X_val)
print("Accuracy:", accuracy_score(y_val, y_pred))
print("\nClassification Report:\n", classification_report(y_val, y_pred))

# Step 8: Process test data in the same way
X_test = test.drop(['Provider', 'PotentialFraud'], axis=1, errors='ignore')  # Exclude Provider and PotentialFraud from features

# Handle missing values for test data
X_test[numeric_cols] = X_test[numeric_cols].fillna(X_test[numeric_cols].mean())
X_test[non_numeric_cols] = X_test[non_numeric_cols].fillna('missing')

# Convert non-numeric columns to string in the test data
for col in non_numeric_cols:
    X_test[col] = X_test[col].astype(str)

# Encode categorical features in the test data using the same LabelEncoders
for col, le in label_encoders.items():
    if col in X_test.columns:
        X_test[col] = le.transform(X_test[col].astype(str))

# Step 9: Make predictions on the test dataset
test_predictions = rf_model.predict(X_test)
print("Test Predictions:", test_predictions)

```
