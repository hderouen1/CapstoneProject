```python
import pandas as pd

# Load the datasets
train = pd.read_csv("train-.csv")
train_inpatient = pd.read_csv("train_inpatientdata.csv")
train_outpatient = pd.read_csv("train_outpatientdata.csv")

### 1️⃣ EDA on `train` dataset
print("🔹 Train Dataset Overview")
print(train.info())  # Structure of data
print(train.describe(include='all'))  # Summary statistics
print("Missing values:\n", train.isnull().sum())
print("Unique Providers:", train["Provider"].nunique())
print("PotentialFraud Distribution:\n", train["PotentialFraud"].value_counts())

### 2️⃣ EDA on `train_inpatient` dataset
print("\n🔹 Train Inpatient Dataset Overview")
print(train_inpatient.info())
print(train_inpatient.describe())
print("Missing values:\n", train_inpatient.isnull().sum())
print("Unique Providers in Inpatient:", train_inpatient["Provider"].nunique())

### 3️⃣ EDA on `train_outpatient` dataset
print("\n🔹 Train Outpatient Dataset Overview")
print(train_outpatient.info())
print(train_outpatient.describe())
print("Missing values:\n", train_outpatient.isnull().sum())
print("Unique Providers in Outpatient:", train_outpatient["Provider"].nunique())

### 4️⃣ Provider Analysis Across Datasets
train_providers = set(train["Provider"].unique())
inpatient_providers = set(train_inpatient["Provider"].unique())
outpatient_providers = set(train_outpatient["Provider"].unique())

# Check overlaps
common_inpatient_train = train_providers.intersection(inpatient_providers)
common_outpatient_train = train_providers.intersection(outpatient_providers)
common_both = inpatient_providers.intersection(outpatient_providers)

print("\n🔹 Provider Analysis")
print(f"Providers in Train only: {len(train_providers - inpatient_providers - outpatient_providers)}")
print(f"Providers in Train & Inpatient: {len(common_inpatient_train)}")
print(f"Providers in Train & Outpatient: {len(common_outpatient_train)}")
print(f"Providers in both Inpatient & Outpatient: {len(common_both)}")
print(f"Providers appearing in ALL datasets: {len(train_providers & inpatient_providers & outpatient_providers)}")
```

```python
import pandas as pd

# Load datasets
train = pd.read_csv("train-.csv")  # Provider-level fraud labels
train_beneficiary = pd.read_csv("train_Beneficiarydata.csv")  # Beneficiary-level details
train_inpatient = pd.read_csv("train_Inpatientdata.csv")  # Inpatient claims
train_outpatient = pd.read_csv("train_Outpatientdata.csv")  # Outpatient claims


# Attempt to parse DOB column correctly
train_beneficiary["DOB"] = pd.to_datetime(train_beneficiary["DOB"], errors="coerce")

# Extract year, then compute age
current_year = 2025
train_beneficiary["Age"] = current_year - train_beneficiary["DOB"].dt.year

# Check if any missing values remain
print(train_beneficiary["Age"].isna().sum())  # Count invalid conversions


for df in [train_inpatient, train_outpatient]:
    if "AdmissionDate" in df.columns and "DischargeDate" in df.columns:
        df["AdmissionDate"] = pd.to_datetime(df["AdmissionDate"], errors="coerce")
        df["DischargeDate"] = pd.to_datetime(df["DischargeDate"], errors="coerce")
        df["DaysAdmitted"] = (df["DischargeDate"] - df["AdmissionDate"]).dt.days
    else:
        print(f"Missing expected date columns in {df}")


# Drop raw date columns after processing
#train_inpatient.drop(columns=["AdmissionDt", "DischargeDt"], inplace=True)
#train_outpatient.drop(columns=["AdmissionDt", "DischargeDt"], inplace=True)

### 2. Merge Beneficiary Data with Claims Using 'BeneID'
train_inpatient = train_inpatient.merge(train_beneficiary, on="BeneID", how="left")
train_outpatient = train_outpatient.merge(train_beneficiary, on="BeneID", how="left")

### 3. Concatenate Inpatient & Outpatient Claims
train_claims = pd.concat([train_inpatient, train_outpatient], ignore_index=True)

### 4. Merge Claims with Main Train Data at the Provider Level
train_full = train_claims.merge(train, on="Provider", how="left")

### 5. Aggregate Features at Provider Level
agg_dict = {
    "InscClaimAmtReimbursed": ["sum", "mean"],  # Total & avg claim amount
    #"AdmissionMonth": ["nunique"],  # Unique admission months per provider
    "ChronicCond_Alzheimer": ["mean"],  # Proportion of Alzheimer's patients
    "ChronicCond_Heartfailure": ["mean"],  # Proportion of heart failure patients
    "ChronicCond_Cancer": ["mean"],  # Proportion of cancer patients
    "RenalDiseaseIndicator": ["mean"],  # Proportion of renal disease cases
    "Age": ["mean", "median"],  # Avg & median patient age per provider
    #"DaysAdmitted": ["sum", "mean"],  # Total & avg length of stay
}

#needs focus dtypes
train_provider_level = train_full.groupby("Provider").agg(agg_dict).reset_index()

# Flatten column names
train_provider_level.columns = ["_".join(col).strip() for col in train_provider_level.columns]

# Merge with fraud labels
train_provider_level = train_provider_level.merge(train[["Provider", "PotentialFraud"]], on="Provider", how="left")

# Encode fraud labels
train_provider_level["PotentialFraud"] = train_provider_level["PotentialFraud"].map({"Yes": 1, "No": 0})

# Save cleaned dataset for modeling
train_provider_level.to_csv("train_processed.csv", index=False)

print("Data processing complete! Ready for model training.")

```
