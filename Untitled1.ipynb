```python
# Summing inpatient claimed amount per provider
inpat_agg = inpat.groupby('Provider').agg({
    'InscClaimAmtReimbursed': 'sum'
}).rename(columns={'InscClaimAmtReimbursed': 'Total_Inpatient_Reimbursed'})

# Summing outpatient claimed amount per provider
outpat_agg = outpat.groupby('Provider').agg({
    'InscClaimAmtReimbursed': 'sum'
}).rename(columns={'InscClaimAmtReimbursed': 'Total_Outpatient_Reimbursed'})

# Merging inpatient and outpatient claims at provider level
provider_claims = inpat_agg.merge(outpat_agg, on='Provider', how='outer').fillna(0)

# Merging with fraud labels
final_df = train.merge(provider_claims, on='Provider', how='left')

# Cross-check: Comparing total amounts with original data
total_inpat_check = inpat['InscClaimAmtReimbursed'].sum()
total_outpat_check = outpat['InscClaimAmtReimbursed'].sum()
total_final_inpat = final_df['Total_Inpatient_Reimbursed'].sum()
total_final_outpat = final_df['Total_Outpatient_Reimbursed'].sum()

print(f"Original Inpatient Total: {total_inpat_check}, Aggregated Inpatient Total: {total_final_inpat}")
print(f"Original Outpatient Total: {total_outpat_check}, Aggregated Outpatient Total: {total_final_outpat}")

# Verifying if numbers match
assert abs(total_inpat_check - total_final_inpat) < 1e-5, "Inpatient totals do not match!"
assert abs(total_outpat_check - total_final_outpat) < 1e-5, "Outpatient totals do not match!"

print("Cross-check passed: Aggregated numbers match original totals!")
```

```python
# Count number of unique patients per provider
unique_patients_per_provider = benef_providers.groupby('Provider')['BeneID'].nunique().reset_index().rename(columns={'BeneID': 'Unique_Patients'})

# Count number of inpatient and outpatient visits per provider
inpat_visits = inpat.groupby('Provider').size().reset_index(name='Inpatient_Visits')
outpat_visits = outpat.groupby('Provider').size().reset_index(name='Outpatient_Visits')

# Average claim amount per visit per provider
inpat_avg_claim = inpat.groupby('Provider')['InscClaimAmtReimbursed'].mean().reset_index().rename(columns={'InscClaimAmtReimbursed': 'Avg_Inpatient_Claim'})
outpat_avg_claim = outpat.groupby('Provider')['InscClaimAmtReimbursed'].mean().reset_index().rename(columns={'InscClaimAmtReimbursed': 'Avg_Outpatient_Claim'})

# Standard deviation of claimed amounts
inpat_std_claim = inpat.groupby('Provider')['InscClaimAmtReimbursed'].std().reset_index().rename(columns={'InscClaimAmtReimbursed': 'Std_Inpatient_Claim'})
outpat_std_claim = outpat.groupby('Provider')['InscClaimAmtReimbursed'].std().reset_index().rename(columns={'InscClaimAmtReimbursed': 'Std_Outpatient_Claim'})

# Merge all new features into final_df
final_df = final_df.merge(unique_patients_per_provider, on='Provider', how='left') \
                   .merge(inpat_visits, on='Provider', how='left') \
                   .merge(outpat_visits, on='Provider', how='left') \
                   .merge(inpat_avg_claim, on='Provider', how='left') \
                   .merge(outpat_avg_claim, on='Provider', how='left') \
                   .merge(inpat_std_claim, on='Provider', how='left') \
                   .merge(outpat_std_claim, on='Provider', how='left')

# Fill NaN values with 0
final_df.fillna(0, inplace=True)
```

```python
# Calculate age as of 2017 (year of the dataset)
benef['DOB'] = pd.to_datetime(benef['DOB'], errors='coerce')
benef['Age'] = 2017 - benef['DOB'].dt.year

# Merge with provider-level information
benef_providers = pd.read_csv('benef_providers.csv')  # Assuming we saved the BeneID-Provider mapping
benef_with_age = benef.merge(benef_providers, on='BeneID', how='inner')

# Aggregate age stats at the provider level
age_stats = benef_with_age.groupby('Provider').agg(
    mean_age=('Age', 'mean'),
    median_age=('Age', 'median'),
    min_age=('Age', 'min'),
    max_age=('Age', 'max')
).reset_index()

# Merge age features into the final dataframe
final_df = pd.read_csv('final_provider_dataset.csv')
final_df = final_df.merge(age_stats, on='Provider', how='left')

# Save the updated dataframe
final_df.to_csv('final_provider_dataset_with_age.csv', index=False)

print("Age feature engineering completed and merged into final_df.")
```

```python
import pandas as pd
import numpy as np



# Identify numeric columns
numeric_cols = final_df.select_dtypes(include=[np.number]).columns

# Apply IQR method for outlier detection and capping
for col in numeric_cols:
    Q1 = final_df[col].quantile(0.25)
    Q3 = final_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap outliers
    final_df[col] = np.clip(final_df[col], lower_bound, upper_bound)

# Save the cleaned dataset
final_df.to_csv('final_provider_dataset_outliers_handled.csv', index=False)

print("Outliers handled and dataset saved as 'final_provider_dataset_outliers_handled.csv'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the original and cleaned datasets
original_df = pd.read_csv('final_provider_dataset.csv')
cleaned_df = pd.read_csv('final_provider_dataset_outliers_handled.csv')

# Identify numeric columns
numeric_cols = original_df.select_dtypes(include=[np.number]).columns

# Plot side-by-side boxplots
plt.figure(figsize=(16, 5 * len(numeric_cols)))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(len(numeric_cols), 2, 2*i - 1)
    sns.boxplot(x=original_df[col], color='lightblue')
    plt.title(f"{col} - Original")

    plt.subplot(len(numeric_cols), 2, 2*i)
    sns.boxplot(x=cleaned_df[col], color='lightgreen')
    plt.title(f"{col} - Cleaned")

plt.tight_layout()
plt.show()
```
